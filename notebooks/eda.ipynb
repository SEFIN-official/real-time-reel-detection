```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/combined_dataset.csv')

# Check balance
print("Class distribution:", df['label'].value_counts())

# Visualize feature distribution by class
df.boxplot(column='packet_size', by='label')
plt.title('Packet Size by Traffic Type')
plt.suptitle('')
plt.show()

df.boxplot(column='duration', by='label')
plt.title('Duration by Traffic Type')
plt.suptitle('')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('../data/combined_dataset.csv')
print(df['label'].value_counts())  # Shows counts of each class
print(df['label'].value_counts(normalize=True))  # Shows proportion of each class
```
